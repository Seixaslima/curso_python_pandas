```python
import pandas as pd

combustiveis_df = pd.read_excel('drive/MyDrive/python-pandas/ca-2021-02.xlsx')
```

```python
#teste se a importação funcional

display(combustiveis_df)
```

```python
#inserção simples de dado

combustiveis_df['Ativo'] = True

display(combustiveis_df.head())
```

```python
# Criar uma coluna "Obs" que tenha nela escrito "MELHOR CIDADE" quando a coluna Municipio for igual a BELO HORIZONTE

combustiveis_df['Obs'] = ['MELHOR CIDADE' if municipio == 'BELO HORIZONTE' else '' for municipio in combustiveis_df['Municipio']]

display(combustiveis_df.loc[combustiveis_df['Municipio'].isin(['BELO HORIZONTE']),['Municipio','Obs']])
```

```python
import numpy as np

combustiveis_df['Valor de Venda - Status'] = np.where(combustiveis_df['Valor de Venda'] > 6.5, 'Tá Caro', 'Tá Barato')
display(combustiveis_df[['Revenda','Valor de Venda','Valor de Venda - Status']])
```

```python
# Calcular postos de gasolina por habitante temos na amostragem de 
# combustiveis nov/2021

num_habitantes_df = pd.read_csv('drive/MyDrive/python-pandas/ibge_num_habitantes_estimado.csv',sep=';')
num_habitantes_df.rename(columns={"Estado":"Estado - Sigla"}, inplace=True)
display(num_habitantes_df)
```

```python
# Faz um MERGE dos dois dataframes

colunas = ['Municipio','Estado - Sigla']

merge_df = combustiveis_df.merge(num_habitantes_df, how='inner', on=colunas)
display(merge_df)
display(merge_df.info())
```

```python
#Destruir coluna completamente vazia (todas as linhas são nulas)

merge_df.dropna(axis='columns', inplace=True)
print(merge_df.info())
```

```python
#Destruindo colunas que não vamos utilizar

colunas=['Regiao - Sigla', 'Nome da Rua', 'Numero Rua', 
         'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda',
         'Unidade de Medida', 'Bandeira', 'Ativo','Valor de Venda - Status','Obs']
merge_df.drop(labels=colunas, axis=1, inplace=True)
print(merge_df.info())
```

```python
# Remover a linhas duplicadas
merge_df.drop_duplicates(inplace=True)
display(merge_df.head(100))
```

```python
#agrupar e contar quantos postos tem na cidade
postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_municipio_df.reset_index(inplace=True)
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_municipio_df.rename(columns={"Revenda": "Numero de Postos"}, inplace=True)
display(postos_por_municipio_df.info())
```

```python
#numero de postos de gasolina por 100mil Habitantes
postos_por_municipio_df['Postos por 100 mil Habitantes'] = (postos_por_municipio_df['Numero de Postos']/postos_por_municipio_df['NumHabitantes2021'])*100000
display(postos_por_municipio_df)
```

```python
#vamos fazer graficos

import matplotlib.pyplot as plt

plt.hist(combustiveis_df['Valor de Venda'])

#colocando Título no grafico
plt.title("Distribuição dos preços dos combustiveis em Novembro de 2021")

#Rótulo horizontal e vertical
plt.xlabel("Preço (em reais)")
plt.ylabel("numero de coletas")

#plotar uma linha vermelha com o preço medio
plt.axvline(combustiveis_df['Valor de Venda'].mean(), color='red', linestyle='dashed', linewidth = 1)
plt.subplots_adjust(bottom=.25, left=.25)
#plotando o grafico
plt.show()
```

```python
c_mean = combustiveis_df['Valor de Venda'].groupby(by=combustiveis_df['Produto']).mean()

display(c_mean)
```

```python
import seaborn as sns
#definindo a area do grafico
plt.figure(figsize=(10,5))


#plotar o grafico
c_mean.plot(
    kind = 'barh',
    xlabel = 'Tipo de combustivel',
    ylabel = 'preço Reais/litro',
    title = 'Media de Preços por combustivel',
    color = 'red',
)

#grid
plt.grid(axis="x")

#remover as bordas superior e direita do grafico
sns.despine()


plt.show()
```

```python
#c_mean.reset_index()

excel = "por_litro.xlsx"
c_mean.to_excel(excel, "Sumario")
```

```python
#Vamor personalizar a planilha Excel...
from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Font, Color, Alignment

#abrindo a planilha excel no openpyxl
wb = load_workbook(excel)

#pegar a planilha correta usando o sheet name (nome da planilha)
ws = wb['Sumario'] #ws = work sheet = planilha em foco

#pintando o cabeçalho de cinza claro
cinzinha = PatternFill("solid",fgColor='cccccc')

coords = ['A1','B1']

for coord in coords:
  ws[coord].fill = cinzinha
  

#onde o combustivel for maior que 6,50  pinta a fonte de vermelho
#e deixa em negrito

MAX_ROW = ws.max_row
num_linha = 2
while (num_linha < MAX_ROW):
  coord = f'B{num_linha}'
  if(ws[coord].value > 6.5):
    ws[coord].font = Font(color='FF0000',bold=True)
  num_linha+=1

#salvar o excel
wb.save(excel)
```
